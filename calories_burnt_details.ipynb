# Import data from calorie csv
Due to potential privacy issues, the file calories_burnt_details.csv is not shared

Filter data

```python
import pandas as pd

calories_burnt_details = pd.read_csv("calories_burnt_details.csv", skiprows=1, index_col=False)
calories_burnt_details.dropna(how="all", axis=1, inplace=True)
calories_burnt_details.columns = [column.replace("com.samsung.shealth.calories_burned.", "") for column in calories_burnt_details.columns]
# calories_burnt_details.drop(columns=["pkg_name", "deviceuuid", "datauuid", "create_time", "update_time", "extra_data"], inplace=True)
```

fix the dates and only select dates from today

```python
calories_burnt_details['day_time'] = pd.to_datetime(calories_burnt_details['day_time'], unit = 'ms')
start_date = pd.Timestamp('2023-08-07')
calories_burnt_details = calories_burnt_details.loc[calories_burnt_details.day_time > start_date]
calories_burnt_details.loc[:, "active_time"] = calories_burnt_details["active_time"] / 60000
calories_burnt_details.set_index("day_time", inplace=True)
calories_burnt_details
```

Check if tef_calorie is significant

```python
calories_burnt_details.loc[calories_burnt_details.tef_calorie != 0.0]
```

Insignificant, so drop

```python
calories_burnt_details.drop(columns=["tef_calorie"], inplace=True)
```

```python
import numpy as np

group_key = np.arange(len(calories_burnt_details)) // 7
calories_burnt_weekly = calories_burnt_details.groupby(group_key).mean()
calories_burnt_weekly.index = [i + 1 for i in range(len(calories_burnt_weekly))]
calories_burnt_weekly
```
